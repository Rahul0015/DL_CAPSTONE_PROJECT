```python
def load_and_merge_data(file1, file2):
    df1 = pd.read_excel(file1, engine='openpyxl')
    df2 = pd.read_excel(file2, engine='openpyxl')
    combined = pd.concat([df1, df2], ignore_index=True)
    return combined

def clean_and_filter_data(df):
    # Filter for India
    df = df[df['country_txt'] == 'India']

    # Select relevant columns
    columns = ['eventid', 'iyear', 'imonth', 'iday', 'country_txt', 'region_txt',
               'provstate', 'city', 'attacktype1_txt', 'targtype1_txt',
               'weaptype1_txt', 'nkill', 'nwound', 'summary']
    df = df[columns]

    # Fill missing values
    df['summary'] = df['summary'].fillna('No summary provided')
    df['nkill'] = df['nkill'].fillna(0)
    df['nwound'] = df['nwound'].fillna(0)

    # Create a new feature: total casualties
    df['casualties'] = df['nkill'] + df['nwound']

    return df

def save_cleaned_data(df, output_path):
    #os.makedirs(os.path.dirname(output_path), exist_ok=True)
    df.to_csv(output_path, index=False)
    print(f"Cleaned data saved to {output_path}")

if __name__ == "__main__":
    # Paths
    file1="C:/Users/rahul/Desktop/Deeplcapstone/data/raw/globalterrordataset1.xlsx"
    file2="C:/Users/rahul/Desktop/Deeplcapstone/data/raw/globalterrordataset2.xlsx"
    output_path = "C:/Users/rahul/Desktop/Deeplcapstone/data/interim/cleaned_india_terrorism.csv" #

    # Process
    df = load_and_merge_data(file1, file2)
    df_clean = clean_and_filter_data(df)
    save_cleaned_data(df_clean, output_path)
   
```
